```python
import numpy as np
import bezier as bz
import matplotlib.pyplot as plt
import nitools as nt
```

```python
A = np.array([[0,0.2,1.0,1.0],[0,0.5,0.2,1.0]])*5
C= bz.Curve(A,3)
C.plot(20)
plt.plot(A[0,:],A[1,:],'k.')
x=np.linspace(0,1,20)
coord = np.zeros((2,20))
for i in range(20):
    coord[:,i:i+1]=C.evaluate(x[i])
plt.plot(coord[0,:],coord[1,:],'r.')
```
